# 02 Cleaning

Objective: build a focused analysis dataset for 30-day readmission using a limited number of high-value columns.

```python
from pathlib import Path
import sys

import pandas as pd

ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    ROOT = ROOT.parent
sys.path.append(str(ROOT / 'src'))

from data_utils import build_clean_dataset, load_mapping_sections
from project_config import IDS_MAPPING_PATH, PROCESSED_DATA_PATH, RAW_DATA_PATH, SELECTED_COLUMNS
```

```python
raw_df = pd.read_csv(RAW_DATA_PATH)
raw_df.shape
```

```python
raw_df[SELECTED_COLUMNS].head()
```

```python
mappings = load_mapping_sections(IDS_MAPPING_PATH)
clean_df = build_clean_dataset(raw_df, mappings)
clean_df.head()
```

```python
missing_summary = clean_df.isna().sum().sort_values(ascending=False)
missing_summary
```

```python
clean_df['readmitted_30'].value_counts(normalize=True).round(4)
```

```python
PROCESSED_DATA_PATH.parent.mkdir(parents=True, exist_ok=True)
clean_df.to_csv(PROCESSED_DATA_PATH, index=False)
print(f'Saved: {PROCESSED_DATA_PATH}')
print(clean_df.shape)
```
